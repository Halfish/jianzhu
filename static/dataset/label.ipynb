```python
# 标签的统计词频
f = open('handcraft_labels.txt', 'wb')
for k, v in sorted(label_count.items(), key = lambda x: x[1], reverse=True):
    string = k + ' ' + str(v) + '\n'
    f.write(string.encode('utf-8'))
f.close()
```
